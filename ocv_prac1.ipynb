# Draw shape & put text

```python
img = cv.imread('data/gohan.jpg')
```

```python
img.shape
```

```python
# BLANK image
blank = np.zeros(img.shape, dtype='uint8')
```

```python
# colour part of image
blank[200:300, 300:400]=200,201,180
plt.imshow(blank)
plt.show()
```

```python
# create rectangel in blacnk image
cv.rectangle(blank, (10,10),(blank.shape[0]//2, blank.shape[1]//2),(0,250,0), thickness=2)
plt.imshow(blank)
plt.show()
```

```python
# draw circle
cv.circle(blank, (blank.shape[0]//2, blank.shape[1]//2), 50, (0,255,255), thickness=3)
plt.imshow(blank)
plt.show()
```

```python
# draw a line
cv.line(blank, (0,0), (blank.shape[0], blank.shape[1]), (0,255,255), thickness = 1)
plt.imshow(blank)
plt.show()
```

```python
# write text on image
cv.putText(blank, 'HIiii', ((blank.shape[0]//2, blank.shape[1]//2)), cv.FONT_HERSHEY_PLAIN, 10.0, (0,255,0), 2)
plt.imshow(blank)
plt.show()
```
