# Edge Detection

## 1. Import Dependencies

```python
import cv2
import os
from matplotlib import pyplot as plt
```

## 2. Load Image

```python
img_path = os.path.join('data','car2.jpg')
```

```python
img = cv2.imread(img_path)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

## 3. Edge Detection

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5,5),0)
canny = cv2.Canny(blur, threshold1 = 50, threshold2=60)
```

```python
plt.imshow(cv2.cvtColor(canny,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# display using opencv
img_rsz = cv2.resize(canny, (int(img.shape[1]/6), int(img.shape[0]/6)))
cv2.imshow('img',img_rsz)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Export result

```python
cv2.imwrite(os.path.join('output','out_car2.jpg'), canny)
```
